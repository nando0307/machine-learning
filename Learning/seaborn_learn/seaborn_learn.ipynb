### Set up

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

### Import data

```python
cs_df = pd.read_csv("./seaborn_data/ComputerSales.csv")
```

```python
cs_df
```

```python
crash_df = sns.load_dataset('car_crashes')
```

```python
crash_df
```

## Distribution plots

### Distribution plot

```python
sns.displot(crash_df['not_distracted'], bins = 5)
```

### Jointplot

```python
sns.jointplot(x = 'speeding', y = 'alcohol', data = crash_df, kind = 'kde')
```

### kde plot

```python
sns.kdeplot(crash_df['alcohol'])
```

### pair plot

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.pairplot(tips_df, hue = 'sex')
```
