### Load packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x = [0, 1, 2, 3, 4]
y = [0, 2, 4, 6, 8]
```

```python
#plt.plot(x, y, label = '2x', c = 'r', linestyle = ':', marker = 'o')
#shorter: [color][marker][line]
plt.figure(figsize=(8, 8), dpi=300)
plt.plot(x, y, "r.:", label = '2x')
#line 2
x2 = np.arange(0,4.5,0.5)
plt.plot(x2, x2**2, label = 'X^2')
plt.xlabel('x')
plt.ylabel('y')

plt.title('My first graph', fontdict = {"fontname":'Comic Sans MS', "fontsize":'20'})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.ylabel('y')

plt.xticks([0, 1, 2, 3])
#plt.yticks([0, 2, 4, 6, 8, 10])
plt.legend()

plt.show()
```
